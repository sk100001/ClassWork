```python
i = sns.load_dataset('iris')
```

```python
i
```

```python
ig = sns.PairGrid(i, hue = "sepal_width")
ig.map(plt.scatter)
```

```python
ig = sns.PairGrid(i, hue = "sepal_width")
ig.map(plt.scatter)
ig.map_diag(plt.hist)
```

```python
ig = sns.PairGrid(i, hue = "sepal_width")
ig.map(plt.scatter)
ig.map_diag(plt.hist)
ig.map_offdiag(plt.scatter)
```

```python
ig=sns.PairGrid(i,hue='species')
ig.map_diag(plt.hist)
ig.map_upper(plt.scatter)
ig.map_lower(sns.kdeplot)
```
